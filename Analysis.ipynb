```python
# ==========================================
# VISUALIZATION
# ==========================================
```

```python
## SALES BY PRODUCTLINE ##
pdt_sales=df.groupby("PRODUCTLINE")["SALES"].sum().sort_values()

plt.figure(figsize=(10,6))
pdt_sales.plot(kind="bar")
plt.title("SALES BY PRODUCTLINE")
plt.xlabel("PRODUCT LINE")
plt.ylabel("SALES")
plt.xticks(rotation=45)
plt.savefig("Images/product_line_sales.png")
plt.show()
```

```python
## SALES BY COUNTRY ##
country_sales=df.groupby("COUNTRY")["SALES"].sum().sort_values()

plt.figure(figsize=(12,5))
country_sales.plot(kind="bar")
plt.title("SALES BY COUNTRY")
plt.xlabel("COUNTRY")
plt.ylabel("TOTAL SALES")
plt.savefig("Images/Country_sales.png")

plt.show()
```

```python
## MONTHLY SALES TREND ##
monthly_sale=df.groupby("MONTH_ID")["SALES"].sum().sort_values()

plt.figure(figsize=(12,6))
monthly_sale.plot(marker="o")
plt.title("MONTHLY SALES TREND")
plt.xlabel("MONTH_ID")
plt.ylabel("TOTAL SALES
plt.savefig("Images/Monthly_sales.png")

plt.show()
```

```python
## TOP 10 CUSTOMER ##
TOP_CUSTOMER=df.groupby("CUSTOMERNAME")["SALES"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
TOP_CUSTOMER.plot(kind="bar")
plt.title("TOP 10 CUSTOMER")
plt.xlabel("CUSTOMER NAME")
plt.ylabel("TOTAL SALES")
plt.show()
```

```python
## SALES BY DEALSIZE ## 
dealsize_sale=df.groupby("DEALSIZE")["SALES"].sum().sort_values()

plt.figure(figsize=(12,5))
dealsize_sale.plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.title("SALES BY DEALSIZE")
plt.savefig("Images/Dealsize_sales.png")

plt.show()
```

```python
## ORDER STATUS ##
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="STATUS")
plt.title("ORDER STATUS DISTRIBUTION")
plt.xticks(rotation=45)
plt.savefig("Images/Order_Status.png")

plt.show()
```

```python
## TOP SELLING PRODUCT ##
top_products = df.groupby("PRODUCTCODE")["SALES"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind="bar")
plt.title("Top Selling Products")
plt.xlabel("Product Code")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.savefig("Images/product_Code_sales.png")

plt.show()
```

```python
## SALES DISTRIBUTION ##
plt.figure(figsize=(10,5))
plt.hist(df["SALES"])
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.savefig("Images/Sales_Distribution.png")

plt.show()
```

```python
## SALES BY TERRITORY ##
territory=df.groupby("TERRITORY")["SALES"].sum()

plt.figure(figsize=(10,4))
territory.plot(kind="bar")
plt.title("Sales by Territory")
plt.xlabel("Territory")
plt.ylabel("Sales")
plt.savefig("Images/Territory_sales.png")

plt.show()
```

```python
## YEAR-WISE SALES ##
year_sales = df.groupby("YEAR_ID")["SALES"].sum()

plt.figure(figsize=(8,5))
year_sales.plot(marker="o")
plt.title("Year-wise Sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.savefig("Images/Yearly_sales.png")

plt.show()
```

```python
## CORRELATION HEATMAP

plt.figure(figsize=(8,6))

sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("Images/Correlation_Heatmap.png")

plt.show()
```

```python
## SALES VS QUANTITY ##

plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="QUANTITYORDERED",y="SALES")
plt.title("Quantity Ordered VS Sales")
plt.savefig("Images/Sales_Vs_Quantity.png")

plt.show()
```
